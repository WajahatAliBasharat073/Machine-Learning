```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
house_price=sklearn.datasets.load_boston()
```

```python
house_df=pd.DataFrame(house_price.data,columns=house_price.feature_names)
```

```python
house_df
```

## add target  to df

```python
house_df['price']=house_price.target
```

```python
house_df
```

```python
## checking num of roes and columns in df
house_df.shape
```

```python
house_df.isnull().sum()
```

```python
## statical measure of dataset
house_df.describe()
```

### Positive correlation

```python
correaltion=house_df.corr()
```

```python
correaltion
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correaltion,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':13},cmap='Blues')
```

```python
x=house_df.drop('price',axis=1)
y=house_df['price']
```

```python
x
```

```python
y
```

### splitting data into train and split data 

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_test.shape
```

```python
x_train.shape
```

## XGBoost Regressor

```python
model=XGBRegressor()
```

```python
model.fit(x_train,y_train)
```

## Evalution
- prediction on training data

```python
## accuracy for prediction on training data
training_data_pred=model.predict(x_train)
```

```python
training_data_pred
```

```python
# R squared error
score_1=metrics.r2_score(y_train,training_data_pred)
print("R squared error : ",score_1)
#Mean Absolute eror
score_2=metrics.mean_absolute_error(y_train,training_data_pred)
print("mean squared error : ",score_2)
```

```python
test_accuracy_pred=model.predict(x_test)
```

```python
test_accuracy_pred
```

```python
# R squared error
score_1=metrics.r2_score(y_test,test_accuracy_pred)
print("R squared error : ",score_1)
#Mean Absolute eror
score_2=metrics.mean_absolute_error(y_test,test_accuracy_pred)
print("mean squared error : ",score_2)
```

```python
## Visulizing the actual and prediction prices
plt.scatter(y_train,training_data_pred)
plt.xlabel("Actual price")
plt.ylabel('predicted price')
plt.title("Actual vs predicted price")
plt.show()
```
